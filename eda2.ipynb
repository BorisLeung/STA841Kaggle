```python
def see_dist(
    data: pd.DataFrame,
    col: str | None = None,
    palette: str = "viridis",
    multiple: str = "layer",
) -> None:
    if col is None:
        sns.displot(
            data,
            x="num_pov",
            stat="density",
            common_norm=False,
            kde=True,
        )
    else:
        print(data[col].value_counts())
        sns.displot(
            data,
            x="num_pov",
            hue=col,
            stat="density",
            common_norm=False,
            kde=True,
            palette=palette,
            multiple=multiple,
        )
```

```python
see_dist(combined_transformed_train_with_num_pov, "edu_q03")
plt.title("attended school vs not")
plt.show()
```

```python
attended_school = combined_transformed_train_with_num_pov[
    combined_transformed_train_with_num_pov["edu_q03"] == 1
]
see_dist(attended_school, "edu_q08")
plt.title("enroll/plan to enroll in school this year vs not")
plt.show()
```

```python
plans_to_attend_school_this_year = attended_school[attended_school["edu_q08"] == 1]
see_dist(plans_to_attend_school_this_year, "edu_q09")
plt.title("attended school this year vs not")
```
